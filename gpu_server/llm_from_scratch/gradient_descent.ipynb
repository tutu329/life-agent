```python
import matplotlib.pyplot as plt
import numpy as np

# 生成x值
x_curve = np.linspace(-2, 2, 400)
x_tangent = np.linspace(0, 2, 200)  # 切线的x范围限制在(0, 2)

# 计算y值
y_curve = x_curve**2

# 计算切线的y值
y_tangent = 2*x_tangent - 1

# 创建图形和轴对象，并调整图形宽高比例
fig, ax = plt.subplots(figsize=(8, 8))  # 调整图形宽高比例

# 绘制曲线
ax.plot(x_curve, y_curve, label='y = x^2')

# 绘制切线
ax.plot(x_tangent, y_tangent, label='Tangent at (1, 1)', linestyle='--')

# 添加标题和标签
ax.set_title('Plot of y = x^2 with Tangent at (1, 1)')
ax.set_xlabel('x')
ax.set_ylabel('y')

# 显示图例
ax.legend()

# 显示网格
ax.grid(True)

# 设置x轴和y轴单位长度相等
ax.set_aspect('equal')

# 显示图形
plt.show()
```
